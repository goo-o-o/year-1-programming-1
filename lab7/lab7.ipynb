1. Slicing and Indexing

data = ["COMP1312", 42, 3.14, ["apple", "banana", "cherry"], "Python", -7]

Write Python expressions (no loops!) from “data” to produce:

1. "COMP1312"
2. ["apple", "banana"]
3. ["banana", "cherry"]
4. ["COMP1312", 42, 3.14]
5. ["COMP1312", "Python", -7]
6. Reverse the entire list using slicing.

````python
data = ["COMP1312", 42, 3.14, ["apple", "banana", "cherry"], "Python", -7]

print(data[:1])
print(data[3][:2])
print(data[3][1:3])
print(data[:3])
print([data[0]] + data[-2:])
print(data[::-1])
````

2. List Mutation and Aliasing

Given the code:

```python
a = [1, 2, 3]
b = a
c = a[:]
d = a.copy()
b.append(99)
c[0] = 100
d[1] = 200
```

1. What are the final contents of a, b, c, and d?
   ```python
   a = [1, 2, 3, 99]
   b = [1, 2, 3, 99]
   c = [100, 2, 3]
   d = [1, 200, 3]

2. Explain why modifying b changes a, but modifying c or d does not.
    - b points to the same object in memory as a, while c and d are shallow copies.
3. Create a multidimensional list and demonstrate a case where slicing/copy still fails (shallow copy).

Example:
```python
m1 = [[1,2], [3,4]]
m2 = m1[:] # shallow copy
```


Modify m2[0][1] and observe both lists.

````python
m1 = [[1,2], [3,4]]
m2 = m1[:] # shallow copy

m2[0][1] = 5678

print(m1)
print(m2)
````

Even though m1 and m2 are different objects in memory, they both point to the same objects in memory that are within their lists, so editing the objects in m2 directly would also edit m1
